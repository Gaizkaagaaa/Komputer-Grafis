```python
import numpy as np
import matplotlib.pyplot as plt

plt.grid(True)
plt.axis('on')

plt.axis([-10, 10, 10, -10])
plt.axes().set_aspect('equal')

x1 = -10
x2 = 10
y1 = 10
y2 = -10
dx = .5
dy = -.5

for x in np.arange(x1, x2, dx):
    for y in np.arange(y1, y2, dy):
        plt.scatter(x, y, s=1, color='lightgray')

plt.plot([-5, 5], [-5, -5], linewidth=2, color='g')
plt.plot([5, 5], [-5, 5], linewidth=2, color='g')

plt.plot([5, -5], [5, 5], linewidth=2, color='y')
plt.plot([-5, -5], [5, -5], linewidth=2, color='y')

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Menentukan batas sumbu grafik
plt.axis([0, 150, 100, 0])

# Menentukan jari-jari lingkaran
r = 40

# Menentukan sudut awal dan akhir
alpha1 = np.radians(0)
alpha2 = np.radians(360)

# Menentukan selisih sudut
dalpha = np.radians(2)

# Menentukan koordinat pusat lingkaran
xc = 75
yc = 50

# Menggambar titik pusat lingkaran
plt.scatter(xc, yc, s=10, color='k')

# Melakukan perulangan untuk menggambar lingkaran
for alpha in np.arange(alpha1, alpha2, dalpha):
    # Menghitung koordinat x dan y pada setiap sudut
    x = xc + r * np.cos(alpha)
    y = yc + r * np.sin(alpha)

    # Menggambar titik pada setiap sudut
    plt.scatter(x, y, s=5, color='k')

# Menampilkan grafik
plt.show()
```

```python
# Mengatur batas sumbu
plt.axis([-10, 140, 90, -10])

# Menyalakan sumbu
plt.axis('on')

# Menampilkan grid
plt.grid(True)

# Menambahkan judul
plt.title('Contoh Sumbu')

# Menentukan langkah grid
dx = 5
dy = -5

# Menggambar garis grid
for x in np.arange(-10, 140, dx):
    for y in np.arange(90, -10, dy):
        plt.scatter(x, y, s=1, color='lightgray')

# Menggambar sumbu koordinat
plt.arrow(0, 0, 20, 0, head_length=4, head_width=3, color='k')
plt.arrow(0, 0, 0, 20, head_length=4, head_width=3, color='k')

# Menampilkan plot
plt.show()
```
